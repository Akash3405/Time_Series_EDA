# Normalizing time series to a Base value (100)

```python
close.head()
```

```python
close.iloc[0] # Checking first row
```

```python
close.div(close.iloc[0])  # Dividing all rows with first row
```

```python
nor_close = close.div(close.iloc[0]).mul(100) # now multiplying each row with 100 
nor_close.head()    
```

```python
nor_close.tail()
```

## So finally the normalized data is ready for comparison and evaluation

```python
plt.style.use("classic")
nor_close.plot(figsize=(12,8))
plt.legend(loc="upper left",fontsize = 18,labelcolor="linecolor")
plt.ylabel("Normalized close price")
plt.title("Normalized close price year wise")
plt.show()
```

<b> Now we have same starting point for all ticker as you can see from above graph,So now we can compare their performance.<b> 
<ul>
<li>As we can see from 2013 to 2018 <b>Microsoft </b>closing pricing increased steadily; after that it was increased very rapidly </li>
<li>For <b>IBM </b> market was going down steadily year by year.</li>
<li>For <b> Boeing Co </b> closing price rises slowly upto 2018 after that it was increased rapidly for 1 year then again it fell down. High fluctations happened in Boeing co from 2013 to 2022.</li>
<li><b> Disney </b> closing price went up and rose slowly upto 2020 mid after that it was increased very rapidly upto 2021 mid then again it fell down. Lot of ups and downs happened in Disney from 2013 to 2022.</li>

## Now checking difference and percentage change in each day for Microsoft

```python
Microsoft = close.MSFT.to_frame()  
Microsoft
```

```python
Microsoft["lag1"] = Microsoft.shift(periods=1)  # shift is basically used for shifitng our data to next cell 
Microsoft
```

```python
Microsoft["Diff"] = Microsoft.MSFT.sub(Microsoft.lag1)
Microsoft
```

```python
Microsoft["%Change"] = Microsoft.MSFT.div(Microsoft.lag1).sub(1).mul(100) # Calculationg % change in each day 
Microsoft
```

```python
Microsoft.MSFT.resample("M").last().pct_change(periods=1).mul(100) # Monthly level % change 
```

 # Return and Risk

```python
ret = close.pct_change().dropna()
ret
```

```python
ret.describe().T.loc[:,["mean","std"]]
```

```python
summary = ret.describe().T.loc[:,["mean","std"]]
summary
```

```python
summary["mean"] = summary["mean"]*261
summary["std"] = summary["std"]*np.sqrt(261)
```

```python
summary
```

```python
plt.style.use("seaborn")
color = ["red","green","yellow","orange"]
summary.plot(kind = "scatter", x = "std", y = "mean", figsize = (15,12), s = 50, fontsize = 15,color = color)
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, "std"]+0.002, summary.loc[i, "mean"]+0.002), size = 35)
plt.xlabel("ann. Risk(std)", fontsize = 15)
plt.ylabel("ann. Return", fontsize = 15)
plt.grid()
plt.title("Risk/Return", fontsize = 20)
plt.show()
```

<ul> Conculsion from the above graph :
<li> In <b> Microsoft </b> we are getting high returns at sustainable risk.</li>
<li> In <b> IBM </b> risk is low but also the return is very less.</li>
<li> In <b> Disney </b> return is lower than Microsoft and Boeing although the risk is sustainable.</li>
<li> In <b> Boeing </b> return is high but at the same time, the risk is very high.
